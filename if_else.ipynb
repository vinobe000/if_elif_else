```python
def num(x):
  if(x>0):
    print("positive")
  elif(x==0):
    print("zero") 
  elif(x<0):
    print("negative")
```

```python
def discount(sales):
  if(sales>3000):
    print(sales*0.20)
  elif(sales>2000):
    print(sales*0.15)
  elif(sales>1000):
    print(sales*0.10)
  else:
    print(sales)
    
```

```python
discount(200)
```

```python
discount(5000)
```

```python
def operators(x,y,z):
  if(z=="+"):
    print(x+y)
  elif(z=="-"):
    print(x-y)
  elif(z=="*"):
    print(x*y)
  elif(z=="**"):
    print(x**y)
  elif(z=="/"):
    print(x/y)
  elif(z=="//"):
    print(x//y)
  else:
    print("operator is not valid")
    
                
```

```python
operators(12,12,"**")
```

```python
operators(12,12,"-")
```

```python
def arithmetic(operators):
  print(operators())
```

```python
def arithmetic(operators,x,y,z):
  return(operators(x,y,z))
```

```python
arithmetic(operators,12,12,"//")
```

```python
arithmetic(operators,12,12,"%")
```
